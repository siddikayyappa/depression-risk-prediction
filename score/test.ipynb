```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/User_scores_nonPII.csv')
df
```

```python
user_ids = list(df['UserId'])[:]
unhealthy_score = df['Unhealthy']
k10_score = df['K-10']
```

```python
for i in range(len(user_ids)):
    user_ids[i] = user_ids[i] + '.csv'
```

```python
dict_1 = dict()
dict_1['userID'] = user_ids
dict_1['unhealthy'] = unhealthy_score
dict_1['k10'] = k10_score
```

```python
pd.DataFrame(dict_1).to_csv('./scores.csv', index=False)
```
